# Monte Carlo Kelly Criterion

```python
import numpy as np
import pandas as pd
from random import random
import matplotlib.pyplot as plt
```

```python
# plt configs
```

```python
# initialised, and rolling variables
time = np.arange(1000)

# now set up the markov chain
ev = 0.02
p = 0.5 + ev
q = 1-p

def single_person_history(bankroll:float,
                       stack_percentage:float,
                       bets:int):
    
    # numpy array for bankroll history
    bankroll_history = np.arange(bets, dtype=float)
    for n in range(bets):
        stack = bankroll*stack_percentage
        bankroll -= stack
        
        # roll the dice
        x = random()
        if x < p:
            # WIN
            stack *=2
            bankroll += stack
        else:
            # hold this L
            stack == 0
        
        bankroll_history[n] = bankroll
        
    return bankroll_history
```

# Monte Carlo

Just run a simple for loop intertwined with matplotlib syntax?

```python
for n in range(20):
    # need to reset these variables every MC
    bankroll = 100
    stack_percentage = 0.01
    bets = 1000
    
    # run 100 simulations
    sim = single_person_history(bankroll,stack_percentage,bets)
    # plot
    plt.plot(time,sim)
    
plt.ylabel('Bankroll')
plt.xlabel('Bets')
plt.show()
```
