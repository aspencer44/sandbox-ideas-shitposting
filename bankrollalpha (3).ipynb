# Bankroll Management: 4 buy-ins, 25% risked every time

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
# Start with 4 buy-ins.
# Each round, AA vs KK; 80:20 favourite
# Strategy: 1/4 of our bankroll every time. 
# Monte Carlo
```

### Whole cell below runs a single MonteCarlo iteration to 1000

```python
# need to initialise every time
stack_percentage = 0.01
odds = 1.9
win_probability = 0.545
print('+/- EV: {}'.format(odds*win_probability-1))

limit = 1000
bankroll = 100
bankroll_history = []
counter = 0
while bankroll > 0:
    while counter != limit:
        counter += 1
    
        # Gamble with stack
        stack = bankroll*stack_percentage
        bankroll -= stack
        random_float = random.random()
        if random_float < win_probability:
            stack *= odds
            bankroll += stack
        else: 
            stack == 0
        bankroll_history.append((counter,bankroll))
        print(counter,bankroll)
    break
```
